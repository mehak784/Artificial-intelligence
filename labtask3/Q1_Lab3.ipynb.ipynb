# Question 1 – Entropy and Information Gain

## Formula:
H(Y) = - Σ p(c) log₂ p(c)

Information Gain:
IG(Y, A) = H(Y) – H(Y | A)

## Calculation (from dataset)
p(Pass) = 3/5  
p(Fail) = 2/5  

H(Result) = –(3/5 log₂ 3/5 + 2/5 log₂ 2/5) ≈ 0.971 bits  

IG(StudyHours) = 0.971  
IG(Attendance) = 0.0199
